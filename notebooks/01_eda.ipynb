# MovieLens EDA
Quick exploratory analysis for the recommendation project.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

root = Path('..').resolve()
raw = root / 'movielens_data'
if not raw.exists():
    raw = root / 'movie_lens_data'

ratings = pd.read_csv(raw / 'ratings.csv')
movies = pd.read_csv(raw / 'movies.csv')
tags = pd.read_csv(raw / 'tags.csv')

ratings.head()
```

```python
summary = {
    'ratings': len(ratings),
    'users': ratings['userId'].nunique(),
    'items': ratings['movieId'].nunique(),
    'tags': len(tags),
}
summary
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='rating', data=ratings)
plt.title('Rating Distribution')
plt.tight_layout()
```

```python
user_counts = ratings.groupby('userId')['movieId'].count()
plt.figure(figsize=(6,4))
sns.histplot(user_counts, bins=50, log_scale=(False, True))
plt.title('Interactions per User')
plt.tight_layout()
```

```python
genres = movies['genres'].str.split('|').explode()
top_genres = genres.value_counts().head(12)
plt.figure(figsize=(8,4))
sns.barplot(x=top_genres.values, y=top_genres.index)
plt.title('Top Genres')
plt.tight_layout()
```

```python
ratings['year'] = pd.to_datetime(ratings['timestamp'], unit='s').dt.year
year_counts = ratings.groupby('year').size()
plt.figure(figsize=(8,4))
year_counts.plot()
plt.title('Ratings Over Time')
plt.tight_layout()
```
